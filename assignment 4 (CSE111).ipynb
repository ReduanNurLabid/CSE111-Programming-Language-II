```python
#task-01
class Customer:
    def __init__(self,name=None):
        self.name = name
    def greet(self,name=None):
        if name == None:
            print("Hello!")
        else:
            print(f"Hello {name}!")
    def purchase(self,*items):
        print(f"{self.name}, you have purchased following item(s):" )
        for i in items:
            print(i)
    
#drivercode
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task-02
class Panda:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender 
        self.age = age
    def sleep(self,duration="unknown"):
        if duration != "unknown":
            if 3 <= duration <= 5:
                food = "Mixed Veggies"
            elif 6<= duration <= 8:
                food = "Eggplant & Tofu"
            elif 9 <=duration <= 11:
                food = "Broccoli Chicken"
            else:
                food = "bamboo leaves"
            return f"{self.name} sleeps {duration} hours daily and should have {food}"
        else:
            return f"{self.name}'s duration is {duration} thus should have only bamboo leaves"

#drivercode
panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
class Student:
    def __init__(self,name="default student"):
        self.name = name
    def quizcalc(self,*score):
        total = 0
        for i in score:
            total += i
        self.avg = total / 3
        
    def printdetail(self):
        print("Hello", self.name)
        print(f"Your average quiz score is {self.avg}")



####################
s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
class Match:
    def __init__(self,teams):
        self.teams = teams.split("-")
        self.runs = 0
        self.overs = 0
        self.wicket = 0
    def print_scoreboard(self):
        print("Batting Team:",self.teams[0])
        print("Bowling Team:",self.teams[1])
        return f"Total Runs: {self.runs} Wickets: {self.wicket} Overs: {self.overs}"
    def add_run(self,runs):
        self.runs += runs
    def add_over(self,overs):
        if self.overs + overs > 5:
            print(f"Warning! Cannot add {overs} over/s (5 over match)")
        else:
            self.overs += overs
    def add_wicket(self,wickets):
        self.wicket = wickets
        
###################
match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(4)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
class Batsman:
    def __init__(self,*args):
        if len(args)==2:
            self.name = "New Batsman"
            self.runs = args[0]
            self.balls = args[1]
    
        else:
            self.name = args[0]
            self.runs = args[1]
            self.balls = args[2]

    def printCareerStatistics(self):
        print(f"Name: {self.name}\nRuns Scored: {self.runs}, Balls Faced: {self.balls}")
    
    def setName(self,name):
        self.name = name
    
    def battingStrikeRate(self):
        self.sr = (self.runs /self.balls)*100
        return self.sr
#########################
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
class TaxiLagbe:
    def __init__(self,car,area):
        self.car = car
        self.area = area

        self.passengers = []
        self.c = 0
    def addPassenger(self,*new_passenger):

        for i in new_passenger:
            if len(self.passengers)<4:
                self.passengers.append(i.split("_"))
                print(f"Dear {self.passengers[self.c][0]}! Welcome to TaxiLagbe.")
                self.c+=1
            else:
                print("Taxi Full! No more passengers can be added.")

        self.total = len(self.passengers)


    def printDetails(self):

        print("Trip info for Taxi number:",self.car)
        print(f"This taxi can cover only {self.area} area.")
        print("Total Passengers:",self.total)

        names =""
        fares =0

        for name,fare in self.passengers:
            names +=name+","
            fares +=int(fare)

        print("Passengers lists:",names[:-1])
        print("Total collected fare:",fares)



#################################

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails() 
```

```python
class Account:
    def __init__(self,name ="Default Account",balance = 0):
        self.name = name
        self.balance = balance
    def details(self):
        print(self.name)
        return self.balance
    def withdraw(self,amount):
        if self.balance-amount<=3070:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
            self.balance -= amount
            print("Withdraw Successful! New balance is:",self.balance)
############################
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
class StudentDatabase:
    def __init__(self,name,roll):
        self.name = name
        self.roll = roll
        self.grades = {}

    def calculateGPA(self,courses,semsester):
        c=[]
        g =[]
        kk={}
        total = 0
        for i in courses:
            c.append(i.split(": "))
        for j,k in c:
            total += float(k)*3
            g.append(j)
        gpa = total/(len(c)*3)
        gpa=round(gpa,2)
        kk[tuple(g)]= gpa
        self.grades[semsester]=kk

    def printDetails(self):
        print("Name:",self.name,"\nID:",self.roll)
        for i,j in self.grades.items():
            print(f"Course Taken in {i}")
            for m in j.keys():
                for p in m:
                    print(p)
            print("GPA:",self.grades[i][m])



        
#######################
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'],
'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'],
'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()


```
