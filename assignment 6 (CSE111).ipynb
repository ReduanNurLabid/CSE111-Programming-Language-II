```python
# 1
class Student:
    serial = 0

    def __init__(self, name, dept, age, cgpa):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa
        Student.serial += 1

    def get_details(self):
        print("ID:", Student.serial)
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)

    @classmethod
    def from_String(cls, str_1):
        name, dept, age, cgpa = str_1.split("-")
        return cls(name, dept, age, cgpa)


s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print("A class variable is a variable decleared in the class, independent of every object in the class.")
print("An instance variable is a variable decleared inside the constructor with self keyword.")
```

```python
# 2
class Assassin:
    no = 0

    def __init__(self, name, rate):
        self.name = name
        self.successRate = rate
        Assassin.no += 1

    def printDetails(self):
        print(
            f"Name: {self.name}\nSuccess rate: {self.successRate}%\nTotal number of Assassin: {Assassin.no}")

    @classmethod
    def failureRate(self, name, rate):
        rate = 100-rate
        return Assassin(name, rate)

    @classmethod
    def failurePercentage(self, name, percentage):
        rate = 100 - int(percentage[:-1])
        return Assassin(name, rate)


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
# 3
class Passenger:
    count = 0
    fare = 450

    def __init__(self, name):
        self.name = name
        Passenger.count += 1

    def set_bag_weight(self, weight):
        self.weight = weight

    def printDetail(self):
        if 21 <= self.weight <= 50:
            fee = 50
        elif self.weight > 50:
            fee = 100
        else:
            fee = 0
        print(f"Name: {self.name}\nBus Fare:{Passenger.fare + fee} taka")


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
class Travel:
    count = 0

    def __init__(self, here, there):
        self.here = here
        self.there = there
        self.time = 1
        Travel.count += 1

    def set_time(self, time):
        self.time = time

    def set_destination(self, there):
        self.there = there

    def set_source(self, here):
        self.here = here

    def display_travel_info(self):
        return f"Source: {self.here}\nDestination:{self.there}\nFlight Time:{self.time}:00"


print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka", "India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur", "Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka", "New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka", "India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller=", Travel.count)
```

```python
# 5
class Employee:
    def __init__(self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod

    @classmethod
    def employeeByJoiningYear(cls, name, year):
        workingPeriod = 2022 - year
        return cls(name, workingPeriod)

    @staticmethod
    def experienceCheck(period, gender):
        if period < 3:
            exp = "not experienced"
        else:
            exp = "experienced"
        if gender.lower() == "male":
            intro = "He"
        else:
            intro = "She"
        return f"{intro} is {exp}"


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Laptop:
    laptopCount = 0

    def __init__(self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopCount += count

    @classmethod
    def resetCount(cls):
        Laptop.laptopCount = 0

    @staticmethod
    def advantage():
        print("Laptops are portable")


lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
class Cat:
    Number_of_cats = 0

    def __init__(self, clr, state):
        self.clr = clr
        self.state = state
        Cat.Number_of_cats += 1

    def printCat(self):
        print(f"{self.clr} is {self.state}")

    def changeColor(self, clr):
        self.clr = clr

    @classmethod
    def no_parameter(cls):
        return cls("White", "sitting")

    @classmethod
    def first_parameter(cls, clr):
        return cls(clr, "sitting")

    @classmethod
    def second_parameter(cls, state):
        return cls("White", state)


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# 8
import math


class Cylinder:
    radius = 5
    height = 18

    def __init__(self, r, h):

        self.r = r
        self.h = h
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        print(f"Updated radius={self.r} and height= {self.h}")
        Cylinder.radius = self.r
        Cylinder.height = self.h

    @classmethod
    def swap(cls, h, r):
        return cls(r, h)

    @classmethod
    def changeFormat(cls, str_1):
        r, h = str_1.split("-")
        return cls(float(r), float(h))

    @staticmethod
    def area(r, h):
        print("Area:", (2*math.pi*r*(r + h)))

    @staticmethod
    def volume(r, h):
        print("Volume:", math.pi*(r**2)*h)


c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```

```python
# 10
class SultansDine:
    count = 0
    sell = 0
    information = {}

    def __init__(self, branch):
        self.branch = branch
        self.branch_sell = 0
        SultansDine.count += 1

    def sellQuantity(self, n):
        if n < 10:
            self.branch_sell = n * 300
        elif n < 20:
            self.branch_sell = n * 350
        else:
            self.branch_sell = n * 400

        SultansDine.sell += self.branch_sell
        SultansDine.information[self.branch] = self.branch_sell

    def branchInformation(self):
        print(
            f"Branch Name: {self.branch}\nBranch Sell: {self.branch_sell} Taka")

    @classmethod
    def details(cls):

        print(
            f"Total number of branch(s): {SultansDine.count}\nTotal Sell:{SultansDine.sell}")
        for branch, branchSell in SultansDine.information.items():
            percentage = (branchSell / SultansDine.sell)*100
            percentage = round(percentage, 2)
            print( 
                f"Branch Name: {branch}, Branch Sell:{branchSell} Taka\nBranch consists of total sell's:{percentage}%")


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
